# Irradiance description
- Light from the source is reflected on the surface just one time and, depending on the slope, the light reflected will or will not be part of the image (i.e. enter the detector)
- Assume the fluctuation of the surface height to be statistically gaussian
- Assume that the surfaces are illuminated by a source, the Sun, of a fixed angular extent $\beta$ are imaged through a lens that subtends a small solid angle
- Assume wind speed is 10m/s

## Details
- Apparent diameter of the sun/light source: $ \beta = 0.68^o$
- glitter function (which describes the sun glint patterns) as a Gaussian function which is in function of $M_x$ and $M_y$ (slope in x and y directions, respectively)
- Source/light Incidence angle (angle between the normal and the source): $\theta_0$
- Detector angle (angle between the normal and the detector): $\theta_d$
- Angle between the x-axis (horizontal) and the surface: $\alpha$
- Anglular subtenses of the source and detector are represented by $\beta$ and $\delta d$, respectively
- In practical situations $\delta d$ << $\beta$, the brightness function (glitter function in one dimension) that operates on the local slope can be approximated by:

$$ B(M) = rect[\frac{M - M_0}{(1+M_0^2)(\beta /2)}], $$

where $ M = \tan{(\alpha)}$ and $M_0 = \tan{[(\theta_0 - \theta_d)/2]} $
- The bidimensional glitter function is defined as:
$$B(M_x,M_y) = \exp{[-\frac{(M_x -\theta_0)^2 + (M_y - \theta_0)^2}{a^2}]} $$

where $a^2$ gives us the width of the Gaussian function, it gives us a physical situation similar to $ \beta$
- $a = M_0 - (1+M_0^2)(\beta /4)$


```python
z_surface = output[:,:,100]*1e6
np.mean(z_surface)
```

```python
def get_image_gradient(img):
    kx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
    ky = np.array([[1,2,1] ,[0,0,0], [-1,-2,-1]])
    # Perform x convolution
    x=scipy.ndimage.convolve(img,kx)
    # Perform y convolution
    y=scipy.ndimage.convolve(img,ky)
    return {'M_x':x,'M_y':y}

M_dict = get_image_gradient(z_surface)
fig, axes = plt.subplots(1,2,figsize=(10,5))
for ax, (title,im) in zip(axes.flatten(),M_dict.items()):
    ax.imshow(im)
    ax.set_title(title)
plt.show()
```
